__B2.__ _(5 points)_ Create a function that takes a year as input and outputs subsets of each of the tables with data from only that year.

```python
def Year_subset(year):
    Batting_year_subset= Batting_keep[Batting_keep["yearID"] == year]
    Fielding_year_subset= Fielding_keep[Fielding_keep["yearID"] == year]
    Pitching_year_subset= Pitching_keep[Pitching_keep["yearID"] == year]
    Salaries_year_subset= Salaries_keep[Salaries_keep["yearID"] == year]
    Teams_year_subset= Teams_keep[Teams_keep["yearID"] == year]
    return [Batting_year_subset,Fielding_year_subset,Pitching_year_subset,Salaries_year_subset,Teams_year_subset]
```

```python
def fix_suffix(In_DF, suffix):
    cols_suffix = In_DF.columns[~In_DF.columns.isin(["playerID","teamID","yearID"])]
    out_df = In_DF.rename(columns = dict(zip(cols_suffix,cols_suffix + "_" + suffix), inplace=False))
    return out_df
```

```python
def baseball_year(year):
    subsetyear = Year_subset(year)
    Batting_df_fix = fix_suffix(subsetyear[0], "Batting")
    Fielding_df_fix = fix_suffix(subsetyear[1], "Fielding")
    Pitching_df_fix = fix_suffix(subsetyear[2], "Pitching")
    Salaries_df_fix = subsetyear[3]
    Table_1 = reduce(lambda left,right: pd.merge(left,right, on = ["playerID","teamID","yearID"]), 
                     [Batting_df_fix,Fielding_df_fix,Pitching_df_fix,Salaries_df_fix])
    Table_2 = pd.merge(Table_1,subsetyear[4], how = "outer", left_on = ["teamID","yearID"], right_on = ["teamID","yearID"])
    Table_final = pd.merge( Master_keep,Table_2,how = "outer",left_on = ["playerID"], right_on = ["playerID"])
    return Table_final
```

```python
test_final_data = baseball_year(2007)
test_final_data.head()
```
